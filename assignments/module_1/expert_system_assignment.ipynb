```python
#print(knowledge_base["nigerian_dish"]["jollof rice"])
desc_ingr = input("(pepper, onion, tomato)").split(",")
print(desc_ingr)
selected_ingr = [f.strip() for f in desc_ingr]
```

```python
#inference_engine

def identify_nigerian_dish(ingredients):
    ingredients = {i.lower().strip() for i in ingredients}
    missing_ingredients = set()
    best_match = None
    max_matches =  0
    

    for nigerian_dish, attributes in knowledge_base["nigerian_dish"].items():
        matched = ingredients.intersection(attributes)
        score = len(matched)

        if score > max_matches:
            max_matches = score
            dish_ingredient = attributes
            best_match = nigerian_dish
            missing_ingredients = set(dish_ingredient) - ingredients

        if best_match:
            if not missing_ingredients:
                return f"you can fully prepare {best_match}"
            else:
                return(
                    f"i think you are preparing {best_match}, "
                    f"but you are missing: {', '.join(missing_ingredients)}"
                )
        #return best_match, missing_ingredients
```

```python
print(identify_nigerian_dish(selected_ingr))
```
