```python
import math
import pandas as pd
```

```python

def area_of_circle(radius):
    area = math.pi * radius ** 2
    return area 

print(area_of_circle(5))

```

```python
def quadratic_equation(a, b, c):
    #x = (-b + math.sqrt(b ** 2 - 4 * a * c)) / 2 * a
    small = (b ** 2) - (4 * a * c)
    x = (-b + math.sqrt(small)) / 2
    return x
print(quadratic_equation(2, 3, 1))
    
```
