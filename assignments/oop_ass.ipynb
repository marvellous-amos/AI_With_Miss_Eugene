```python
#user profile
class UserProfile:
    def __init__(self, user_name, dietary_preference, weekly_food_budget):
        self.user_name = user_name
        self.dietary_preference = dietary_preference
        self.weekly_food_budget = weekly_food_budget

file1 = open("user_profile.txt", "w")

def save_to_file(self, filename= "user_profile.txt"):
    with open(filename, "w") as file1:
        file1.write(self.user_name + "\n")
        file1.write(self.dietary_preference + "\n")
        file1.write(str(self.weekly_food_budget + "\n"))

def load_from_file(cls, filename= "user_profile.txt"):
    try:
        with open(filename, "r") as file1:
            user_name = file1.readline().strip()
            dietary_preference = file1.readline().strip()
            weekly_food_budget = float(file1.readline().strip())
            return cls(user_name, dietary_preference, weekly_food_budget)
    except FileNotFoundError:
        return None
    except ValueError:
        print("Error loading profile data.")
        return None        
```
