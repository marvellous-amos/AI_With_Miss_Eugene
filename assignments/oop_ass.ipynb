```python
#user profile
class UserProfile:
    def __init__(self, user_name, dietary_preference, weekly_food_budget):
        self.user_name = user_name
        self.dietary_preference = dietary_preference
        self.weekly_food_budget = weekly_food_budget

    def __str__(self):
        return f"user_name: {self.user_name}\ndietary_preference: {self.dietary_preference}\nweekly_food_budget: {self.weekly_food_budget}"

#create an instance
user = UserProfile("Ibinabo", "Vegan", 50000)

#save to a text file
with open("user_profile.txt", "w") as file1:
    file1.write(str(user))
```

```python
#ingredient storage
class Ingredient:
    def __init__(self, name, quantity):
        self.name = name
        self.quantity = quantity

    def __str__(self):
        return f"name: {self.name}\nquantity: {self.quantity}"
    
#create an instance
jollof_rice = Ingredient(
    name= ["rice", "tomatoes", "oil", "onions"],
    quantity= ["4 cups", "2 tins", "1 litre", "2 bulbs"]
)

#save to a text file
with open("jollof_rice.txt", "w") as file2:
    file2.write(str(jollof_rice))


```
