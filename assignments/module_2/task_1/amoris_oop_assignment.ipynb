```python
class Ingredient:
    def __init__(self, Name, Quantity):
        self.Name = Name
        self.Quantity = Quantity
```

```python
class User_Profile:
    def __init__(self, name, dietary_preference, weekly_budget):
        self.name = name
        self.dietary_preference = dietary_preference
        self.weekly_budget = weekly_budget
```

## Functional R1

```python
profileFile="user_profile.txt"

with open(profileFile, "r") as file:
    content = file.read()

if content == "":
    print("No user profile found. Please enter your details.")
    with open(profileFile, "w") as file:
        user = User_Profile(input("Enter your name: "), input("Enter your dietary preference: "), input("Enter your weekly budget: "))
        user_details = f"{user.name}\n{user.dietary_preference}\n{user.weekly_budget}\n"
        file.write(user_details)
        print("User profile created successfully.")
else:
    print("User profile found:")
    print(content)
```

```python
ingredientFile="ingredients.txt"

def add_ingredient(name, quantity):
    with open(ingredientFile, "a") as file:
        ingredient = Ingredient(name, quantity)
        ingredient_details = f"{ingredient.Name},{ingredient.Quantity}\n"
        file.write(ingredient_details)
        print(f"Added ingredient: {ingredient.Name} with quantity: {ingredient.Quantity}")
```

```python
ingredients_list = []
num_ingredients = int(input("How many ingredients do you want to add? "))
for n in range(num_ingredients):
    name = input("Enter ingredient name: ")
    quantity = input("Enter ingredient quantity: ")
    ingredients_list.append((name, quantity))

for name, quantity in ingredients_list:
    add_ingredient(name, quantity)

with open(ingredientFile, "r") as file:
    ingredients_content = file.read()
```
