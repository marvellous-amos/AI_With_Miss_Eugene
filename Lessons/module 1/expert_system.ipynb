```python
#problem_memory

auxilary_features = {
    "1": ["has hair", "gives milk"],
    "2": ["eats meat", "has sharp teeth", "has claws", "has forward looking eyes"],
    "3": ["has hooves", "chews cuds"]
}
primary_features = {
    "mammal": auxilary_features["1"],
    "carnivore": auxilary_features["2"],
    "ungulate": auxilary_features["1"] + auxilary_features["3"]
}
secondary_features = ["red-brown color", "dark spots", "dark stripes"]

animal_knowledge_base = {
    "animals": {
        "monkey": {
            "primary features": primary_features["mammal"],
            "secondary features": [secondary_features[0] + secondary_features[1]]
        },
        "tiger": {
            "primary features": primary_features["carnivore"],
            "secondary features": secondary_features[1]
        },
        "giraffe": {
            "primary features": primary_features["ungulate"],
            "secondary features": secondary_features[2]
        },
        "zebra": {
            "primary features": primary_features["ungulate"],
            "secondary features": secondary_features[2]
        },
    },
}

#print(animal_knowledge_base["animals"]["zebra"])
desc_pri_feat = input("(has hooves, chews cud, has hair, gives milk, eats meat)").split(",")
desc_sec_feat = input("(red brown color, dark spots, dark stripes) select those that apply separated by commas").split(",")
#print(desc_pri_feat + desc_sec_feat)
selected_feat = [f.strip() for f in desc_pri_feat + desc_sec_feat]

```

```python
print(f"you have selected the following features: {selected_feat}")

#inference_engine

def identify_animal(features):
    features = set(features)
    best_match = "unknown animal"
    max_matches = 0

    for animal, attributes in animal_knowledge_base["animals"].items():
        primary = set(attributes["primary features"])
        secondary = set(attributes["secondary features"])

        primary_matches = len(features & primary)
        secondary_matches = len(features & secondary)
        total_matches = primary_matches + secondary_matches

        if primary_matches >= len(primary) / 2:
            if total_matches > max_matches:
                max_matches = total_matches
                best_match = animal
    return best_match

print(f"you have selected the following features: {selected_feat}")

```

```python
print(identify_animal(selected_feat))
```
