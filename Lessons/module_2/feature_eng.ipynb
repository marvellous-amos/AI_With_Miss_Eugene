```python
numerical_columns = df.select_dtypes("number").columns
print(numerical_columns)
```

```python
for column in numerical_columns:
    median = df[column].median()
    df[column].fillna(median, inplace=True)
```

```python
df.info()
```

```python
year_columns = [feature for feature in numerical_columns if 'Yr' in feature or 'Year' in feature]
print(year_columns)
for column in year_columns:
    df[column] = df['YrSold'] - df[column]
```

```python
df["YearBuilt"] = df["YrSold"] - df["YearBuilt"]
```

```python
df["YearBuilt"].head(5)
```

```python
df["RemodAge"] = df["YrSold"] - df["YearRemodAdd"]
```

```python
numerical_0s = df.loc[:, (df == 0).any()].select_dtypes(include=['number']).columns
print(numerical_0s)
numerical_columns = numerical_columns.difference(numerical_0s)
```

```python
print(numerical_columns)
```

```python
skewness = df[numerical_columns].skew()
skew_column = skewness[abs(skewness) > 1]
print(skew_column)
```

```python
skew_features = ["GrLivArea", "LotArea", "LotFrontage", "MSSubClass", "SalePrice"]
for column in skew_features:
    df[column] = np.log(df[column])
```

```python
categorical_columns = df.select_dtypes(include=['object', 'category']).columns
# categorical_columns = [
#     col for col in df.columns
#     if df[col].dtype == 'object'
#]
print(categorical_columns)
for column in categorical_columns:
    labels_ordered = df.groupby([column])['SalePrice'].mean().sort_values().index
    labels_ordered = {x: i for i, x in enumerate(labels_ordered, 0)}
    df[column] = df[column].map(labels_ordered)
```
